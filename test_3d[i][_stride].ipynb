```python
import numpy as np
from numpy import random
```

```python
element_1_2d = np.random.randint(0, 9, (20, 3, 2))
element_2_2d = np.random.randint(0, 9, (21, 3, 2))
element_3_2d = np.random.randint(0, 9, (22, 3, 2))
element_4_2d = np.random.randint(0, 9, (19, 3, 2))
element_5_2d = np.random.randint(0, 9, (18, 3, 2))

test_2d = []
test_2d.append(element_1_2d)
test_2d.append(element_2_2d)
test_2d.append(element_3_2d)
test_2d.append(element_4_2d)
test_2d.append(element_5_2d)
```

```python
element_1_3d = np.random.randint(0, 9, (20, 3, 3))
element_2_3d = np.random.randint(0, 9, (21, 3, 3))
element_3_3d = np.random.randint(0, 9, (22, 3, 3))
element_4_3d = np.random.randint(0, 9, (19, 3, 3))
element_5_3d = np.random.randint(0, 9, (18, 3, 3))

test_3d = []
test_3d.append(element_1_3d)
test_3d.append(element_2_3d)
test_3d.append(element_3_3d)
test_3d.append(element_4_3d)
test_3d.append(element_5_3d)
```

```python
stride = 3
```

```python
# Downsample as requested
for i in range(len(test_2d)):
    test_2d[i] = test_2d[i][::stride]  # get every strideth value, if stride == 2
    if test_3d is not None:
        test_3d[i] = test_3d[i][::stride]
```

```python
# 处理后的数据
print('test_3d[0].shape : ', test_3d[0].shape)
print(test_3d[0])
```

```python
# 原数据
print('element_1_3d.shape : ', element_1_3d.shape)
print(element_1_3d)
```
