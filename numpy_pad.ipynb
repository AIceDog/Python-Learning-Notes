```python
import numpy as np
```

```python
test = np.zeros((5, 7, 2))
print(test)
```

```python
print('test.shape', test.shape)
print('test.shape[0] : ', test.shape[0])
print('test.shape[1] : ', test.shape[1])
print('test.shape[2] : ', test.shape[2])
```

```python
test = np.zeros((5, 7, 2))

# 对 test 进行 padding
# 在 shape[0] 上方填充值 1，下方填充值 2
# 在 shape[1] 上方填充值 3，下方填充值 4
# 在 shape[2] 上方填充值 5，下方填充值 6

# 在 shape[0] 上方填充 1 次，下方填充 2 次
test_shape0 = np.pad(test, ((1, 2), (0, 0), (0, 0)), 'constant', constant_values=((1, 2), (3, 4), (5, 6)))

print('test_shape0.shape : ', test_shape0.shape)
print('test_shape0 : ', test_shape0)
```
